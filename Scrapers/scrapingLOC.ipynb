#### Extract Movies from Library of Congress

```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
import re
```

```python
# Create a list to store the data
data = []
```

```python
r = requests.get("https://www.loc.gov/programs/national-film-preservation-board/film-registry/complete-national-film-registry-listing/")
soup = BeautifulSoup(r.text, 'html.parser')
titles = soup.find_all('th', {'scope': 'row'})
for title in titles:
    year = title.find_next_sibling('td')
    if year is None:
        continue
    if re.match(r'^\d{4}$', year.text):
        print(title.text, year.text)
    else:
        continue
```

```python
# Extract the title and year and add them to the data list
for title in titles:
    movie_title = title.text.strip()
    year = title.find_next_sibling('td')
    if year is None:
        continue
    if re.match(r'^\d{4}$', year.text):
        movie_year = year.text
    else:
        continue
        
    data.append([movie_title, movie_year])
```

```python
data.append([movie_title, movie_year])
```

```python
# Define the filename for the .csv file
filename = "loc.csv"
```

```python
# Write the data to the .csv file
with open(filename, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['title', 'year'])  # Write header row
    writer.writerows(data)  # Write movie data

print("Data saved to", filename)
```
