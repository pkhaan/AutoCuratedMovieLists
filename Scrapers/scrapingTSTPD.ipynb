```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import re
```

```python
r = requests.get("https://www.theyshootpictures.com/gf1000_all1000films.htm")
soup = BeautifulSoup(r.text, 'html.parser')
movies = soup.find('div', id='stacks_in_1784')

#Get the text of the second span tag
movie_list = movies.find_all('span')[1:2][0]

#Replace all <br> tags with \n
for br in movie_list.find_all("br"):
    br.replace_with("\n")

#Split the text into a list of movies and reove empty strings
movie_list_list = list(filter(None, movie_list.text.split("\n")))

for movie in movie_list_list:
    print(movie)
```

Now we want to extract the movie, title, year and director from each line and add them to a Pandas DataFrame

```python
extracted = []
for movie in movie_list_list:
    #movie_name = re.search(r'(\s([A-Za-z]+\s)+)', movie)
    movie_name = re.search(r'(.*)\)(.*)\((.*),\s([0-9]+)', movie)
    extracted.append(
        {'Title:' : movie_name.group(2),
         'Director:' : movie_name.group(3),
         'Year:' : movie_name.group(4)
        }
    )

df = pd.DataFrame(extracted)
df.head(40)

title_array = []
for title in df['Title:']:
    title_array.append(title)

print(title_array[:50])
```
